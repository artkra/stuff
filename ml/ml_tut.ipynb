```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.svm import SVC
```

```python
n = 1024
X = np.random.normal(2.5,0.7,n)
Y = np.random.normal(2.5,0.7,n)
# T = (X-2.5)**2 + (Y-2.5)**2 < 0.7
T = X + Y < 5
plt.axes([1,1,1,1])
plt.scatter(X,Y, s=75, c=T, cmap='coolwarm', alpha=.5)

plt.xlim(0,5)
plt.ylim(0,5)
# savefig('../figures/scatter_ex.png',dpi=48)
plt.show()
```

```python
data = np.vstack((X, Y)).T
data.shape
```

```python
data_train, data_test, y_train, y_test = train_test_split(data, T)
```

```python
ppn = Perceptron(alpha=0.001, n_iter=100, eta0=0.5)
svc = SVC(kernel='rbf', random_state=0, gamma=0.20, C=15.0)
```

```python
ppn.fit(data_train, y_train)
svc.fit(data_train, y_train)
```

```python
ppn.predict((2.5,2.2))
```

```python
def plot_decision_regions(X, y, classifier, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    edgecolor='black',
                    marker=markers[idx], 
                    label=cl)
```

```python
plot_decision_regions(data, T, classifier=ppn)
plt.legend(loc='upper left')
plt.title('Perceptron Accuracy: {}'.format(ppn.score(data_test, y_test)))
plt.tight_layout()
# plt.savefig('./perceptron_2.png', dpi=300)
plt.show()
```

```python
n = 1024
X = np.random.normal(2.5,0.7,n)
Y = np.random.normal(2.5,0.7,n)
TT = (X-2.5)**2 + (Y-2.5)**2 < 1
plt.axes([1,1,1,1])
plt.scatter(X,Y, s=75, c=TT, cmap='coolwarm', alpha=.5)

plt.xlim(0,5)
plt.ylim(0,5)
# savefig('../figures/scatter_ex.png',dpi=48)
plt.show()
```

```python
data = np.vstack((X, Y)).T
data.shape
```

```python
data_train, data_test, y_train, y_test = train_test_split(data, TT)
```

```python
svc = SVC(kernel='rbf', random_state=0, gamma=0.20, C=15.0)
svc.fit(data_train, y_train)
```

```python
plot_decision_regions(data, TT, svc)
plt.title('SVM with kernel trick. Accuracy: {}'.format(svc.score(data_test, y_test)))
plt.legend()
# plt.savefig('svm_with_rbf.png')
plt.show()
```
